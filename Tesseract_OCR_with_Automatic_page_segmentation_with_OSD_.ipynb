```python
import os
import glob

# Folder path
folder_path = "/content/invoice/images"

# Patterns to match
patterns = ["*_thresholded.*", "*_resized.*","*_enhanced.*","*_enhanced_rescaled.*","*_rescaled.*"]

# Loop through patterns and remove matching files
for pattern in patterns:
    # Create the full pattern path
    full_pattern = os.path.join(folder_path, pattern)
    # Get the list of files matching the pattern
    files_to_remove = glob.glob(full_pattern)
    for file_path in files_to_remove:
        try:
            os.remove(file_path)  # Remove the file
            print(f"Removed: {file_path}")
        except Exception as e:
            print(f"Error removing {file_path}: {e}")
```

```python
import pytesseract
import cv2
import numpy as np
import os

# Set the path to the image folder and the output folder
image_folder = "/content/invoice/images"  # Path to folder containing images
output_folder = "/content/extracted_texts/"  # Path to save extracted texts

# Create output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Set Tesseract executable path
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

# Counter to keep track of the number of processed images
image_count = 0
max_images = 30 # Limit the number of images to process

# Loop through all images in the image folder
for filename in os.listdir(image_folder):
    if filename.endswith(".jpg") or filename.endswith(".png"):  # Adjust image formats if needed
        # Stop after processing the maximum number of images

        # Read the image
        img_path = os.path.join(image_folder, filename)
        img = cv2.imread(img_path)

        # Resize the image to enhance OCR accuracy (optional)
        img = cv2.resize(img, None, fx=2, fy=2)

        # Convert the image to grayscale
        #img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Apply dilation (optional, enhances text connections)
        #kernel = np.ones((1, 1), np.uint8)
        #img = cv2.dilate(img, kernel, iterations=1)


        config = '--psm 1'

        extracted_text = pytesseract.image_to_string(img, config=config, lang='eng')

        # Save the extracted text to a file
        output_text_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.txt")
        with open(output_text_path, 'w', encoding='utf-8') as text_file:
            text_file.write(extracted_text)

        # Print the extracted text (optional)
        print(f"Extracted text from {filename}:\n{extracted_text[:500]}...\n")  # Print first 500 characters

        # Increment the image counter
        image_count += 1
```
